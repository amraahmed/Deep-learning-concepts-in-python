```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
################################# Sodtmax ##################################

# Generating Random integers

z = np.random.randint(-5,high=10,size=15)
print(z)

#Computing Softmax

num = np.exp(z)
den = np.sum(num)
sig = num/den
print(sig)

#Compare
plt.plot(z,sig,'ko')
plt.xlabel("Original number (z)")
plt.yscale('log')
plt.ylabel('Softmax $sig$')
plt.title('$\sum\sigma$=%g' %np.sum(sig))
plt.show()
```

```python
## Using Instance
softmx = nn.Softmax(dim=0)
sigmaT = softmx(torch.Tensor(z))
print(sigmaT)
plt.plot(z,sig,'ko')
plt.xlabel("Original number (z)")
plt.yscale('log')
plt.ylabel('Softmax $sig$')
plt.title('$\sum\sigma$=%g' %np.sum(sig))
plt.show()
```

```python
# logarithms
x = np.linspace(0.1,1,30)
logx = np.log(x)
fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':15})
plt.plot(x,logx,'ks-',markerfacecolor='w')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.show()
```

```python
################################## Entropy ##################################
# Probability of something
P = [.25,.75]
H = 0
for p in P:
    H += -p*np.log(p)
    
print(f"Entropy is {str(H)}")
```

```python
## Cross Entropy
p = [.9,0.1]
q = [.3,.7]
H = 0
for i in range(len(p)):
    H += -(q[i]*np.log(p[i]))
print(f"Cross entropy is {str(H)}")
```

```python
### Torch Entropy ###

p_tensor = torch.Tensor(p)
q_tensor = torch.Tensor(q)

F.binary_cross_entropy(p_tensor,q_tensor)
```
